##1.2 Arrays
Programs on 1-d arrays like - sum of elements of array, searching an element inarray, finding minimum and maximum element in array, count the number ofeven and odd numbers in array. For all such programs, also find the timeComplexity, compare if there are multiple methods

```python
def sum_arr(arr):
  # This function calculates the sum of all elements in an array.
  # Time Complexity: O(N), where N is the number of elements in the array.
  # The function iterates through each element once.
  total = 0
  for num in arr:
    total += num
  return total
arr =  [12, 3, 4, 15,45,78,450]
print(sum_arr(arr))
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr):
  # This function finds the maximum (or minimum) element in an array.
  # Time Complexity: O(N), where N is the number of elements in the array.
  # The function iterates through each element once.
  max_num = arr[0]
  for num in arr:
    if num > max_num:
      max_num = num
  return max_num
arr =  [12, 3, 4, 15,45,78,450]
print(max_min(arr))
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(arr):
  even_count = 0
  odd_count = 0
  for num in arr:
    if num % 2 == 0:
      even_count += 1
    else:
      odd_count += 1
  return even_count, odd_count
arr =  [12, 3, 4, 15,45,78,450]
print(Counting_EvenOdd(arr))
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr, target):
  low = 0
  high = len(arr) - 1
  while low <= high:
    mid = (low + high) // 2
    if arr[mid] == target:
      return mid
    elif arr[mid] < target:
      low = mid + 1
    else:
      high = mid - 1
      return -1
arr = [1, 3, 5, 7, 9]
print(Binary_Search(arr, 5))
```
